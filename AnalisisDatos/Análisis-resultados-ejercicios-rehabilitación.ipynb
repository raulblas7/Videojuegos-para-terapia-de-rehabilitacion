### El notebook busca enseñar al fisioterapeuta los resultados obtenidos por el paciente durante la partida de uno de los videojuegos jugados

Es por ello que deberá introducir en la celda siguiente el nombre del archivo de texto que quiere analizar, dicho archivo es **imprescindible** que esté ubicado al mismo nivel que dicho notebook, para que este pueda leerlo sin ningún tipo de fallo.

```python
archivo = input('Ingresa el nombre del archivo: ')
```

### Gráfica de puntos

En la siguiente gráfica el fisioterapeuta podrá observar en un gráfico 2D los valores obtenidos en el movimiento del paciente en los ejes X e Y, además de ver a través de una escala de colores (*turbo*) los valores en Z.

```python
import matplotlib.pyplot as plt
import numpy as np

# Primero vamos a quitarles los paréntesis al archivo y crear uno nuevo sin ellos
with open(archivo, 'r') as file:
    text = file.read()

text = text.replace('(', '').replace(')', '')

with open('archivo_sin_parentesis.txt', 'w') as file:
    file.write(text)

# Lee el archivo de texto y almacena sus valores en una lista
data = np.loadtxt('archivo_sin_parentesis.txt', delimiter=',')

# Separa los valores de la lista en tres listas distintas
x = data[:,0]
y = data[:,1]
z = data[:,2]

# Crea una figura y un eje
fig, ax = plt.subplots()

# Grafica los puntos de la lista que corresponden a los dos primeros valores de cada línea del archivo
ax.scatter(x, y, c=z, cmap='turbo')

# Agrega etiquetas a los ejes y un título a la gráfica
ax.set_xlabel('Valor x')
ax.set_ylabel('Valor y')
ax.set_title('Gráfica de puntos')

# Muestra la gráfica
plt.show()
```

### Gráfica 3D (a modo mapa de calor)

En la siguiente gráfica el fisioterapeuta podrá observar en un gráfico 3D los valores obtenidos en el movimiento del paciente en los ejes X, Y y Z, y donde se utiliza la misma escala de colores que en la gráfica anterior.

```python
from mpl_toolkits.mplot3d import Axes3D

# Crea la figura y el eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_trisurf(x, y, z, cmap='turbo', edgecolor='none')

# Ajustamos la vista de la figura y agregamos una etiqueta al eje z
ax.view_init(30, 45)
ax.set_zlabel('Valor z')
ax.set_xlabel('Valor x')
ax.set_ylabel('Valor y')

# Agregamos una barra de colores
fig.colorbar(surface, shrink=0.5, aspect=5)

# Mostramos la figura
plt.show()
```

### Gráfica 3D interactiva

En la siguiente gráfica el fisioterapeuta podrá interactuar para así observar más detenedimente los resultados obtenidos por el paciente.

```python
import plotly.graph_objs as go

# Crear una figura
fig = go.Figure()

# Añadir un gráfico de puntos en 3D
fig.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=5, color=z, colorscale='Turbo', opacity=0.8)))

# Añadir etiquetas de ejes y título
fig.update_layout(scene=dict(xaxis_title='Eje X', yaxis_title='Eje Y', zaxis_title='Eje Z'), title='Gráfico 3D interactivo')

# Mostrar la figura
fig.show()
```
